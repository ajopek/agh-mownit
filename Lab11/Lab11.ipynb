```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

## Równania atraktora Rössler'a:

<math> \begin{cases}  \frac{dx}{dt} = -y - z \\ \frac{dy}{dt} = x + ay \\ \frac{dz}{dt} = b + z(x-c) \end{cases} </math>

```python
# Constants in Rossler equations
a = 0.2
b = 0.2
c = 5.7

# Rossler equations
def rossler(state, t):
  x, y, z = state
  return (-y - z), x + a * y, b + z*(x - c)
```

```python
#Calculate states
init_state = [1.0, 1.0, 1.0]
t = np.arange(0.0, 100.0, 0.01)

states = odeint(rossler, init_state, t)
```

## Wykresy dla różnych wartości parametrów

```python
# Plot
fig = plt.figure()
axis = fig.gca(projection='3d')
axis.plot(states[:,0], states[:,1], states[:,2])
plt.show()
```
